## (A) Identification of Candidate Datasets

### Dataset 1: BatteryLife – A Comprehensive Dataset and Benchmark for Battery Life Prediction

- **Dataset name and source**:  
  BatteryLife Dataset, introduced in the KDD 2025 Datasets and Benchmarks Track paper *“BatteryLife: A Comprehensive Dataset and Benchmark for Battery Life Prediction”*. Dataset and code are publicly available via the authors’ GitHub repository.

- **Course topic alignment**:  
  Time series mining, anomaly detection, and regression analysis on sequential data.

- **Potential beyond-course techniques**:  
  Deep learning–based time series forecasting (e.g., LSTM, Transformer models), autoencoder-based anomaly detection, representation learning for battery degradation.

- **Dataset size and structure**:  
  Aggregation of 16 battery datasets containing approximately 90,000 time-series samples from around 1,000 batteries. Data consists of multivariate sequences with varying lengths across charge–discharge cycles.

- **Data types**:  
  Time series measurements (voltage, current, capacity), cycle index, timestamps, and battery metadata (chemistry, temperature, operating conditions).

- **Target variable(s), if any**:  
  Battery life or remaining useful life (continuous regression target).

- **Licensing or usage constraints**:  
  Publicly available for research use; distributed under an open research-friendly license via GitHub and Zenodo.

### Dataset 2: IrrMap – A Large-Scale Comprehensive Dataset for Irrigation Method Mapping

- **Dataset name and source**:  
  IrrMap Dataset, introduced in the KDD 2025 paper *“IrrMap: A Large-Scale Comprehensive Dataset for Irrigation Method Mapping”*. Dataset and code are available via GitHub and HuggingFace.

- **Course topic alignment**:  
  Clustering, spatial data mining, and feature-based classification.

- **Potential beyond-course techniques**:  
  Deep learning for image classification and segmentation (CNNs, Vision Transformers), multi-modal data fusion, and spatial pattern mining.

- **Dataset size and structure**:  
  Over 1 million satellite image patches (224×224 pixels) collected from multiple geographic regions. Data is organized as labeled image samples with accompanying metadata.

- **Data types**:  
  Multi-band satellite imagery (raster data), vegetation indices, geographic coordinates, crop and land-use attributes.

- **Target variable(s), if any**:  
  Irrigation method category (e.g., flood, sprinkler, drip), treated as a supervised classification target.

- **Licensing or usage constraints**:  
  Open-access dataset released for research purposes, with code and data loaders provided by the authors.

### Dataset 3: Instance-Dependent Label Noise Graph Benchmark

- **Dataset name and source**:  
  Instance-Dependent Label Noise Graph Dataset, introduced in the KDD 2025 paper *“Delving into Instance-Dependent Label Noise in Graph Data: A Comprehensive Study and Benchmark”*. Dataset and benchmark code are released alongside the paper.

- **Course topic alignment**:

- **Potential beyond-course techniques**:  
  Graph Neural Networks (GNNs), noise-robust learning, self-supervised graph representation learning.

- **Dataset size and structure**:  
  Multiple graph datasets with thousands to millions of nodes and edges, augmented with controlled instance-dependent label noise for benchmarking.

- **Data types**:  
  Graph structures (nodes and edges), node features, clean and noisy node labels.

- **Target variable(s), if any**:  
  Node classification labels, with the presence of synthetic or real label noise.

- **Licensing or usage constraints**:  
  Publicly available for academic research; usage governed by the dataset license specified in the accompanying repository.

## (D) Exploratory Data Analysis (Selected Dataset Only)


### Distribution of Cycle Life

```python
import json
import pandas as pd

with open("Dataset/Life labels/total_MICH_labels.json", "r") as f:
    mich_labels = json.load(f)

# Convert to DataFrame
life_df = pd.DataFrame(
    list(mich_labels.items()),
    columns=["battery_id", "cycle_life"]
)

print(life_df.head())
print("Number of MICH batteries:", len(life_df))

import matplotlib.pyplot as plt

plt.figure()
plt.hist(life_df["cycle_life"], bins=20)
plt.xlabel("Cycle Life")
plt.ylabel("Frequency")
plt.title("Distribution of Cycle Life (MICH)")
plt.show()

```

```python
print(life_df["cycle_life"].describe())
```

### Sequence Length Distribution

```python
import os
import pandas as pd

mich_path = "Dataset/MICH"

lengths = []

for file in os.listdir(mich_path):
    if file.endswith(".pkl"):
        df = pd.read_pickle(os.path.join(mich_path, file))
        lengths.append(len(df))

length_df = pd.DataFrame(lengths, columns=["sequence_length"])

print(length_df.describe())

import matplotlib.pyplot as plt

plt.figure()
plt.hist(length_df["sequence_length"], bins=20)
plt.xlabel("Number of Time Steps")
plt.ylabel("Frequency")
plt.title("Distribution of Sequence Lengths (MICH)")
plt.show()
```

### Initial capacity distribution

```python
initial_caps = []

for file in os.listdir(mich_path):
    battery = pd.read_pickle(os.path.join(mich_path, file))
    first_cycle = battery['cycle_data'][0]
    initial_caps.append(max(first_cycle['discharge_capacity_in_Ah']))

plt.figure()
plt.hist(initial_caps, bins=20)
plt.xlabel("Initial Discharge Capacity (Ah)")
plt.ylabel("Frequency")
plt.title("Distribution of Initial Capacities")
plt.show()
```

### Capacity degradation curve

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

battery = pd.read_pickle(os.path.join(mich_path, os.listdir(mich_path)[0]))
cycles = battery['cycle_data']

capacities = []

for cycle in cycles:
    discharge_cap = cycle['discharge_capacity_in_Ah']
    capacities.append(max(discharge_cap))

plt.figure()
plt.plot(capacities)
plt.xlabel("Cycle Number")
plt.ylabel("Discharge Capacity (Ah)")
plt.title("Capacity Degradation Curve")
plt.show()
```

## (E) Initial Insights and Direction

Key Observations from EDA

1. Significant Variability in Cycle Life
Cycle life varies widely across MICH batteries, indicating heterogeneous degradation behavior.
→ Models must generalize across diverse aging patterns.

2. Variable-Length Time Series
The number of cycles per battery differs substantially.
→ Modeling approaches must handle variable-length sequential data.

3. Nonlinear Capacity Degradation
Capacity fade is nonlinear, with slow early degradation followed by accelerated decline near end-of-life.
→ Linear models may be insufficient to capture this behavior.

4. Variation in Initial Capacity
Initial capacities differ across batteries, suggesting manufacturing and operating variability.
→ Normalization and representation learning may improve robustness.

Project Direction

Based on these insights, this project will:

Implement classical time-series or regression baselines (course-aligned methods).

Compare them with a deep sequence model (e.g., LSTM or Transformer) as a beyond-course technique.

Evaluate whether nonlinear models better capture degradation dynamics and improve life prediction accuracy.

## Algorithmic Decision Justification

Target: Predict Remaining Useful Life (RUL).
Why: Cycle life is continuous and degradation is nonlinear, so predicting RUL captures progression better than classification.

Features: Early-cycle capacity and voltage curves.
Why: EDA showed early degradation patterns correlate with long-term failure.

Normalization: Normalize by initial capacity.
Why: Initial capacities vary across batteries; normalization isolates degradation behavior.

Model Choice:

Linear Regression (baseline) — interpretable reference model.

LSTM (advanced) — handles nonlinear temporal sequences observed in degradation curves.

Train/Test Split: Split by battery, not by cycle.
Why: Prevents data leakage from future cycles of the same battery.

Metric: MAE/RMSE.
Why: Suitable for continuous RUL prediction and easy to interpret.
